```python
import numpy as np
import pandas as pd
df=pd.read_csv('Laptop.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.groupby('Model').count()['Rating']
```

```python
num_rating_df=df.groupby('Model').count()['Rating'].reset_index()
num_rating_df.rename(columns={'Ratings':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_ratings_df=df.groupby('Model').mean().reset_index()
avg_ratings_df.rename(columns={'Ratings':'avg_ratings'},inplace=True)
avg_ratings_df
```

```python
popular_df=num_rating_df.merge(avg_ratings_df,on='Model')
popular_df
```

```python
popular_df=popular_df.drop('Number',axis=1)
popular_df
```

```python
popular_df=popular_df.drop('RAM',axis=1)
popular_df
```

```python
popular_df=popular_df.drop('Price',axis=1)
popular_df
```

```python
popular_df[popular_df['Rating_y']>=3.1].sort_values('Rating_x',ascending=False).head(20)
```

```python
popular_df=popular_df[popular_df['Rating_y']>=3.1].sort_values('Rating_x',ascending=False).head(20)
popular_df
```
